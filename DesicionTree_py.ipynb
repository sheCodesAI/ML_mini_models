```python
import pandas as pd
df = pd.read_csv("/content/salaries (2).csv")
df.head()
```

```python
 inputs = df.drop('salary_more_then_100k',axis='columns')
 target = df['salary_more_then_100k']
 inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
# Converts text → numbers
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

ex : create a new encoder(converter) for company column

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()
```

👉 Take company column
👉 Convert text → numbers
👉 Store in new column company_n

fit_transform 👉 do 2 things:

learn text values

convert into numbers

```python
# drop those label cols
inputs_n = inputs.drop(['company', 'job', 'degree'], axis='columns')
inputs_n
```

train our classifier now

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2)
```

# ***Ready to train our classifier***

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```

```python
model.predict([[2,2,1]])
```

```python
prediction_input = pd.DataFrame([[2, 2, 1]], columns=['company_n', 'job_n', 'degree_n'])

# Now, predict using this DataFrame
model.predict(prediction_input)
```
